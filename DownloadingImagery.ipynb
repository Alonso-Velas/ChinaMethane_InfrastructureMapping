```python
import pandas as pd
import numpy as np
import geopandas as gpd
import leafmap
from leafmap import tms_to_geotiff
import localtileserver
import xarray
from tqdm import tqdm
```

```python
# Load input GeoDataFrame
ordos1_gdf = gpd.read_file("Data/Individual_Scene_Facilities/ordos1_industrial.gpkg",
                                     layer = "ordos1_industrial_dissolved_attributes")

##Convert CRS to UTM 49 for area calculation
ordos1_gdf = ordos1_gdf.to_crs('EPSG:32649')

ordos1_gdf['area']=ordos1_gdf['geometry'].area/1000000

# Reproject to WGS84 
ordos1_gdf84 = ordos1_gdf.to_crs('EPSG:4326')
```

```python
# Loop through each row in the GeoDataFrame
for index, row in ordos1_gdf84.iterrows():
    # Extract bounds for the current geometry [minx, miny, maxx, maxy]

    # ordos1_gdf.at[index, 'searched'] = 1
    bbox = row['geometry'].bounds

    bbox_list = list(bbox)
    
    # Define unique filename based on index
    #change prefix based on region
    image_name = rf'C:/Users/alons/OneDrive - Duke University/EDF_ChinaMethane/Data/Imagery/ordos1/ordos1_{index}.tif'
    
    # Download/save the imagery, zoom 18 is about 0.6m resolution, which is good for identifying infrastructure
    tms_to_geotiff(
        output=image_name, 
        bbox=bbox_list, 
        zoom=18, 
        source='https://server.arcgisonline.com/ArcGIS/rest/services/World_Imagery/MapServer/tile/{z}/{y}/{x}', 
        overwrite=False
    )
    print(f'Processed {image_name}')
```
